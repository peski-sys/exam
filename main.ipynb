```python
def has_coconut(strings):
    """Check if 'coconut' is in the list of strings."""
    return "coconut" in strings
# Test the function
print(has_coconut(["apple", "banana", "coconut"]))
print(has_coconut(["apple", "banana", "orange"]))
print(has_coconut(["Coconut", "banana"]))
print(has_coconut(["coconut"]))
```

```python
strings = []

while True:
    user_input = input("Enter a string (press Enter to finish): ")
    if user_input == "":
        break
    strings.append(user_input)

# Convert to lowercase and sort alphabetically
lowercase_strings = [s.lower() for s in strings]
lowercase_strings.sort()

# Display the sorted strings
print("\nStrings in lowercase, alphabetical order:")
for s in lowercase_strings:
    print(s)
```
